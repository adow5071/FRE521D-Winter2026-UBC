### 2.4 Connect Git to GitHub (SSH Key Setup)

SSH keys let you connect to GitHub without entering your password each time.

---

#### Step 1: Generate SSH Key

Open terminal and run:

```bash
# Generate a new SSH key (replace with your GitHub email)
ssh-keygen -t ed25519 -C "your.email@student.ubc.ca"

# When prompted:
# - Press Enter to accept the default file location
# - Enter a passphrase (optional but recommended) or press Enter for none
```

---

#### Step 2: Add SSH Key to SSH Agent

**Windows (Git Bash):**
```bash
# Start the SSH agent
eval "$(ssh-agent -s)"

# Add your key
ssh-add ~/.ssh/id_ed25519
```

**macOS:**
```bash
# Start the SSH agent
eval "$(ssh-agent -s)"

# Add your key to the agent and keychain
ssh-add --apple-use-keychain ~/.ssh/id_ed25519
```

---

#### Step 3: Copy Your Public Key

**Windows:**
```bash
# Display the public key, then copy it
cat ~/.ssh/id_ed25519.pub
```

**macOS:**
```bash
# Copy directly to clipboard
pbcopy < ~/.ssh/id_ed25519.pub
```

---

#### Step 4: Add Key to GitHub

1. Go to GitHub.com and sign in
2. Click your profile picture (top right) → **Settings**
3. In the left sidebar, click **SSH and GPG keys**
4. Click **New SSH key**
5. Title: "My Laptop" (or something descriptive)
6. Key type: **Authentication Key**
7. Paste your public key in the "Key" field
8. Click **Add SSH key**

---

#### Step 5: Test Connection

```bash
ssh -T git@github.com
```

You should see: `Hi username! You've successfully authenticated...`

Open PowerShell as Administrator and run:

**Step 2: Install Docker Desktop**

1. Download from: **[https://www.docker.com/products/docker-desktop](https://www.docker.com/products/docker-desktop)**

### 4.3 Create Course Docker Configuration

We will use Docker Compose to set up MySQL. Create a project folder and configuration file.

**Step 1: Create course folder**

```bash
# Create a folder for the course
mkdir fre521d
cd fre521d
```

**Step 2: Create docker-compose.yml file**

Create a file named `docker-compose.yml` in your course folder with the following content:

````python
# Save this content as docker-compose.yml in your fre521d folder

docker_compose_content = '''
version: "3.8"

services:
  mysql:
    image: mysql:8.0
    container_name: fre521d_mysql
    restart: unless-stopped
    environment:
      MYSQL_ROOT_PASSWORD: rootpassword
      MYSQL_DATABASE: mfre521d
      MYSQL_USER: mfre521d_user
      MYSQL_PASSWORD: mfre521d_user_pw
    ports:
      - "3306:3306"
    volumes:
      - mysql_data:/var/lib/mysql
      - ./data:/data
    command: --default-authentication-plugin=mysql_native_password

volumes:
  mysql_data:
'''

print("Copy the content above (without the triple quotes) and save it as docker-compose.yml")
````

You can also create this file using the terminal:

**Windows (PowerShell):**
```powershell
# Navigate to your course folder first, then run:
notepad docker-compose.yml
# Paste the content and save
```

**macOS/Linux:**
```bash
# Navigate to your course folder first, then run:
nano docker-compose.yml
# Paste the content, then Ctrl+O to save, Ctrl+X to exit

# Or use VS Code:
code docker-compose.yml
```

### 4.4 Start MySQL Container

Navigate to your course folder and run:

```bash
# Start MySQL container (first time will download the image)
docker compose up -d

# Check that container is running
docker ps
```

You should see `fre521d_mysql` in the container list.

### 4.5 MySQL Connection Details

Use these settings to connect to MySQL from Python or database tools:

| Setting | Value |
|---------|-------|
| Host | `localhost` or `127.0.0.1` |
| Port | `3306` |
| Database | `mfre521d` |
| Username | `mfre521d_user` |

### 4.6 Test MySQL Connection

````python
# Test MySQL connection from Python
# Make sure Docker is running and the container is up!

import mysql.connector

try:
    connection = mysql.connector.connect(
        host='127.0.0.1',
        port=3306,
        database='mfre521d',
        user='mfre521d_user',
        password='mfre521d_user_pw'
    )
    
    if connection.is_connected():
        cursor = connection.cursor()
        cursor.execute("SELECT VERSION()")
        version = cursor.fetchone()[0]
        print(f"Connected to MySQL Server version: {version}")
        cursor.close()
        connection.close()
        print("MySQL connection test: SUCCESS")
        
except Exception as e:
    print(f"MySQL connection test: FAILED")
    print(f"Error: {e}")
    print("\nTroubleshooting:")
    print("1. Is Docker Desktop running?")
    print("2. Is the MySQL container running? (docker ps)")
    print("3. Did you start the container? (docker compose up -d)")
````

<h2 style="color: #002145;">7. Troubleshooting Common Issues</h2>

### Anaconda / Python

**"conda is not recognized" (Windows)**
- Reinstall Anaconda and check "Add to PATH"
- Or open "Anaconda Prompt" instead of regular Command Prompt

**"conda: command not found" (macOS)**
- Run `conda init zsh` (or `conda init bash`)
- Close and reopen Terminal

### Git / GitHub

**SSH key not working**
- Ensure SSH agent is running: `eval "$(ssh-agent -s)"`
- Re-add key: `ssh-add ~/.ssh/id_ed25519`
- Check key is on GitHub: Settings → SSH keys

**"Permission denied (publickey)"**
- Verify key exists: `ls -la ~/.ssh/`
- Check correct key is added to GitHub
- Try regenerating the key

### Docker

**Docker Desktop won't start (Windows)**
- Ensure WSL 2 is installed: `wsl --status`
- Enable virtualization in BIOS
- Restart computer

**"Cannot connect to Docker daemon"**
- Make sure Docker Desktop is running (check system tray/menu bar)
- Restart Docker Desktop

**MySQL container not starting**
- Check logs: `docker logs fre521d_mysql`
- Port conflict: Check if another MySQL is running on port 3306
- Remove old container: `docker compose down -v` then `docker compose up -d`

### VS Code

**"code: command not found" (macOS)**
- Open VS Code
- Open Command Palette (`Cmd+Shift+P`)
- Type "Shell Command: Install 'code' command in PATH"

**Python interpreter not found**
- Open Command Palette → "Python: Select Interpreter"
- If fre521d not listed, restart VS Code
- Verify environment exists: `conda env list`
